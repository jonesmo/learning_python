```
#Is there a statistically significant difference between the prices of houses that have an open porch vs not?
#How do you handle null values?
#will need to transform variable openporchsf so it can be used in a t test
from scipy import stats

open_porches = (houseprices_df['openporchsf'] > 0)
open_porches_df = houseprices_df.iloc[open_porches.values]
no_open_porch = (houseprices_df['openporchsf'] == 0)
no_open_porch_df = houseprices_df.iloc[no_open_porch.values]

#Neither DF has nulls in the saleprice column
#The columns aren't the same length, so have to take a sample of open_porches_df that's the same length as no_open_porch_df
print(open_porches_df.shape[0])
print(no_open_porch_df.shape[0])

#I'm just taking the first 656 listed homes from open_porches_df
open_porches_df = open_porches_df.iloc[0:656]

stats.pearsonr(open_porches_df['saleprice'], no_open_porch_df['saleprice'])
#The correlation between prices of houses with and without open porches is almost zero, and the p-value is high.
#This calculation doesn't support a statistically significant difference in prices of houses based on whether they have an open porch.
```

```
#Is there a significant correlation between lotarea and saleprice?  Positively or negatively correlated?
stats.pearsonr(houseprices_df['lotarea'], houseprices_df['saleprice'])

#In this case, there is a positive correlation between lotarea and saleprice
#with a p-value very close to zero.  Bigger lots correlate with higher price.
```
