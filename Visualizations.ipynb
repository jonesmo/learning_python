```
#import everything
!pip install seaborn == 0.9.0

import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')

houseprices_df = pd.read_csv('/content/gdrive/My Drive/Thinkful/python_for_data_scientists/houseprices.csv')
```

```
#illustrate the distribution of saleprice
plt.figure(figsize=(12, 5))
price_hist = sns.distplot(houseprices_df['saleprice'], bins = 65, kde=False)
```

```
#illustrate the distribution of lotarea
plt.figure(figsize=(12, 5))
plt.xlim(0, 35000)
lotarea_hist = sns.distplot(houseprices_df['lotarea'], bins=150, kde=False)
```

```
#use a boxplot to illustrate how overallcond affects the sale price
plt.figure(figsize=(10,10))
condition = sns.boxplot(x = 'overallcond', y = 'saleprice', data=houseprices_df);
```

```
#on the same boxplot, highlight the difference between having central air conditioning or not
plt.figure(figsize=(15,10))
plt.ylim(0, 650000)
condition = sns.boxplot(x = 'overallcond', y = 'saleprice', hue='centralair', data=houseprices_df);
condition.set(xlabel='Overall Condition', ylabel='Sale Price');
```

```
#is there a significant correlation between lotarea and the saleprice? Are they positively or negatively correlated?
from scipy import stats
plt.figure(figsize=(15,10))
plt.xlim(0, 40000)
lot_vs_price = sns.scatterplot(x='lotarea', y='saleprice', data=houseprices_df);

stats.pearsonr(houseprices_df['lotarea'], houseprices_df['saleprice'])

#Because the p-value of the correlation is close to zero, there is a significant positive correlation between lot area and sale price.
```
