```
#NOW ON TO CENSUS ASSIGNMENT

census_population_df = pd.read_csv('/content/gdrive/My Drive/Thinkful/python_for_data_scientists/state-populations.csv')
census_regions_df = pd.read_csv('/content/gdrive/My Drive/Thinkful/python_for_data_scientists/census-divisions.csv')

#Merge the two DFs into one called census

census = pd.merge(census_population_df, census_regions_df)
census.head()
```

```
#Re-shape census so that one column contains all population measures and another the year attributes
census.head()
census_melted = pd.melt(frame=census, id_vars=['region','division','state'], var_name='year', value_name='population')
census_sorted = census_melted.sort_values(by=['state','year'])
census_sorted.head(20)
```

```
#Group the data by year and summarize it
census_summarized = census_sorted.groupby(['year']).sum()
census_summarized = census_summarized.assign(average_per_state = census_summarized.population / 51)
census_summarized.head()
```

```
#Group the data by region, division, and year and summarize it
census_grouped = census_sorted.groupby(['region','division','year']).describe()
census_grouped.head(20)
```
